# DARTH VADER

```python
%pylab inline
from scipy.fftpack import fft, fftfreq
import scikits.audiolab as audiolab
```

## Análisis de voz de Darth Vader

```python
#Conversión del archivo de entrada a un canal

#Se lee el archivo de audio y se guardan en arreglos la señal, la tasa de muestreo y la codificación del archivo
entrada1_DV, tasa_muestreo1_DV, enc1_DV = audiolab.wavread("Darth_Vader.wav")
#Darth_Vader.wav es de dos canales iguales, por lo que se toma sólo el primero y se guarda en el archivo Darth_Vader_entrada.wav
y1_DV=entrada1_DV[:,0]
format = audiolab.Format('wav')
salida = audiolab.Sndfile('Darth_Vader_entrada.wav','w',format,1,tasa_muestreo1_DV)
salida.write_frames(y1_DV.real)
salida.close()
```

```python
#Se lee el archivo de un canal
entrada_DV, tasa_muestreo_DV, enc_DV = audiolab.wavread("Darth_Vader_entrada.wav")
#Normalización de la señal de entrada
y_DV=entrada_DV/max(entrada_DV)
#Definición de arreglo de tiempo
t_DV = arange(0, len(y_DV)/float(tasa_muestreo_DV), 1/float(tasa_muestreo_DV))
#Número de datos
n_DV = len(y_DV)
#delta de tiempo
dt_DV=1/float(5000)
```

```python
#Gráfica de la señal de entrada
plot(t_DV, y_DV)
xlabel("Tiempo(s)")
ylabel("Amplitud normalizada")
title('Voz de Darth Vaden')
show()
```

Se observa un retardo más o menos constante en tiempo entre los picos , es decir, espacios con amplitudes nulas

```python
#Tranformada de Fourier
Y_DV=fft(y_DV)
#Arreglo con la frecuencia de cada punto de la señal
frecuencias_DV=fftfreq(n_DV,dt_DV)
```

```python
#Espectro en frecuencia normalizado de la señal DV
plot(abs(frecuencias_DV),abs(Y_DV)/max(Y_DV[1:]))
axis([1,1000,0,1.1])
xlabel('Frecuencia (Hz)')
ylabel('Valor absoluto de la amplitud normalizada')
title('ESPECTRO EN FRECUENCIA DV')
show()
```

Se observa que las frecuencias con mayores amplitudes se encuentran alrededor de los 280Hz, siendo importante el rango aproximado de 40Hz a 840Hz 

```python
#Obtención de la frecuencia principal de la señal

#Posiciones de los puntos cuyas frecuencias tienen mayor amplitud, ordenadas de manera descendente
c = (argsort(-abs(Y_DV)))
print 'La frecuencia con mayor amplitud es:',round(abs(frecuencias_DV[c[1]]))
```

# Archivo de entrada

```python
#Se lee el archivo de audio y se guardan en arreglos la señal, la tasa de muestreo y la codificación del archivo
entrada1, tasa_muestreo1, enc1 = audiolab.wavread("serrucho.wav")
#El archivo serrucho.wav es de dos canales iguales, por lo que se toma sólo el primero y se guarda en el archivo serrucho1.wav
y1=entrada1[:,0]
format = audiolab.Format('wav')
salida = audiolab.Sndfile('serrucho1.wav','w',format,1,tasa_muestreo1)
salida.write_frames(y1.real)
salida.close()
```

```python
#Se lee el archivo de un canal
entrada, tasa_muestreo, enc = audiolab.wavread("serrucho1.wav")
#Normalización de la señal de entrada
y=entrada/max(entrada)
#Número de datos
n = len(y)
#Definición de arreglo de tiempo
t = arange(0, n/float(tasa_muestreo), 1/float(tasa_muestreo))
#delta de tiempo
dt=1/float(5000)
```

```python
#Gráfica de la señal de entrada
plot(t, y)
xlabel("Tiempo(s)")
ylabel("Amplitud normalizada")
title('SENIAL DE ENTRADA')
show()
```

```python
#Tranformada de Fourier
Y=fft(y)
#Arreglo con la frecuencia de cada punto de la señal
frecuencias=fftfreq(n,dt)
```

```python
#Espectro en frecuencia de la señal serrucho
subplot(1,2,1)
plot(abs(frecuencias),abs(Y)/max(Y[1:]))
axis([1,1000,0,1.1])
xlabel('Frecuencia (Hz)')
ylabel('Valor absoluto de la amplitud normalizada')
title('SERRUCHO')
#Espectro en frecuencia de la señal DV
subplot(1,2,2)
plot(abs(frecuencias_DV),abs(Y_DV)/max(Y_DV[1:]))
axis([1,1000,0,1.1])
xlabel('Frecuencia (Hz)')
title('DARTH VADER')
show()
```

Se observa que los espectros en frecuencia son muy distintos

# Filtro pasa banda

```python
#Arreglo que va a guardar los puntos filtrados en el dominio de la frecuencia
Y_filtrada=zeros(n)
#Se aplican los factores que relacionan las amplitudes del archivo de entrada con las de Darth Vader
#para los rangos de frecuencia identificados
#Para las demás frecuencias se deja la amplitud original
for i in range(n):
    if(frecuencias[i]> 40 and frecuencias[i]<100):
        Y_filtrada[i]=Y[i]*0.5
    elif(frecuencias[i]> 100 and frecuencias[i]<200):
        Y_filtrada[i]=Y[i]*3.0
    elif(frecuencias[i]> 200 and frecuencias[i]<250):
        Y_filtrada[i]=Y[i]*6.0
    elif(frecuencias[i]> 250 and frecuencias[i]<320):
        Y_filtrada[i]=Y[i]*10.0
    elif(frecuencias[i]> 320 and frecuencias[i]<350):
        Y_filtrada[i]=Y[i]*8.0
    elif(frecuencias[i]> 350 and frecuencias[i]<370):
        Y_filtrada[i]=Y[i]*3.0
    elif(frecuencias[i]> 370 and frecuencias[i]<400):
        Y_filtrada[i]=Y[i]*1.0
    elif(frecuencias[i]> 400 and frecuencias[i]<550):
        Y_filtrada[i]=Y[i]*3.0
    elif(frecuencias[i]> 550 and frecuencias[i]<720):
        Y_filtrada[i]=Y[i]*9.0
    elif(frecuencias[i]> 720 and frecuencias[i]<840):
        Y_filtrada[i]=Y[i]*5.0
    else:
        Y_filtrada[i]=Y[i]
```

```python
#Espectro en frecuencia de la señal serrucho filtrada
subplot(1,3,2)
plot(abs(frecuencias),abs(Y_filtrada/max(Y_filtrada[1:])))
axis([1,1000,0,1.4])
xlabel('Frecuencia (Hz)')
title('serrucho')
#Espectro en frecuencia de la señal de Darth Vader
subplot(1,3,3)
plot(abs(frecuencias_DV),abs(Y_DV)/max(Y_DV[1:]))
axis([1,1000,0,1.4])
xlabel('Frecuencia (Hz)')
title('Darth Vaden')
#Espectro en frecuencia de la señal serrucho de entrada
subplot(1,3,1)
plot(abs(frecuencias),abs(Y)/max(Y[1:]))
axis([1,1000,0,1.4])
xlabel('Frecuencia (Hz)')
ylabel('Valor absoluto de la amplitud normalizada')
title('serrucho original')
show()
```

Se observa que el espectro en frecuencia se asemeja considerablemente en su forma al de Darth Vader

Para el caso de un archivo de entrada arbitrario se podría simplemente filtrar dentro del rango de frecuencias principales de Darth Vader sin modificar las amplitudes. Esto no daría un espectro en frecuencias tan cercano pero sería una buena aproximación

```python
#Transformada inversa de Fourier
y_filtrada=ifft(Y_filtrada)
#Retardo en el tiempo
y_final=zeros(4*len(y_filtrada))
for i in range(len(y_filtrada)-2):
    y_final[4*i]=y_filtrada[i]
    y_final[(4*i)+1]=y_filtrada[i+1]
    y_final[(4*i)+2]=y_filtrada[i+2]
#Se genera un archivo wav que guarda la señal filtrada con una tasa de muestreo mayor, relacionada con el retardo 
format = audiolab.Format('wav')
salida = audiolab.Sndfile('serrucho_Vader.wav','w',format,1,tasa_muestreo*3)
salida.write_frames(y_final.real)
salida.close()
```

```python
#Gráfica de la señal de entrada y la señal filtrada (sin retardo)
subplot(2,1,1)
plot(t,y,c='b')
axis([0,6.5,-0.2,1.3])
xlabel('Tiempo (s)')
ylabel('Amplitud normalizada')
title('serrucho original')
subplot(2,1,2)
plot(t,y_filtrada,c='k')
axis([0,6.5,-0.2,1.3])
ylabel('Amplitud normalizada')
title('serrucho filtrado')
show()
```

La ganancia observada se debe a los factores utilizados para filtrar la señal
