# Filtro Pasa Banda

```python
%pylab inline
from scipy.fftpack import fft, fftfreq
import scikits.audiolab as audiolab
```

```python
#Se lee el archivo de audio y se guardan en arreglos la señal, la tasa de muestreo y la codificación del archivo
entrada, tasa_muestreo, enc = audiolab.wavread("chimpanzee.wav")
#Normalización de la señal de entrada
y=entrada/max(entrada)
#Definición de arreglo de tiempo
t = arange(0, len(y)/float(tasa_muestreo), 1/float(tasa_muestreo))
#Número de datos
n = len(y)
#delta de tiempo
dt=1/float(5000)
```

```python
#Gráfica de la señal de entrada
plot(t, y)
xlabel("Tiempo(s)")
ylabel("Amplitud normalizada")
```

```python
#Tranformada de Fourier
Y=fft(y)
#Arreglo con la frecuencia de cada punto de la señal
frecuencias=fftfreq(n,dt)
```

```python
#Obtención de la frecuencia principal de la señal

#Posiciones de los puntos cuyas frecuencias tienen mayor amplitud, ordenadas de manera descendente
c = (argsort(-abs(Y)))
print 'Frecuencias con mayores amplitudes:',abs(frecuencias[c[:10]])
```

```python
#Se observa que el valor en la posición 3 es el que tiene sentido físico
#puesto que los anteriores son casi cero
freq=abs(frecuencias[c[3]])
print 'La frecuencia principal de la señal es aproximadamente', round(freq),'Hz'
```

```python
#Filtro pasa banda

#Definición de la banda de frecuencias a filtrar
frec_min=700
frec_max=1000
#Arreglo que va a guardar los puntos filtrados en el dominio de la frecuencia
Y_filtrada=zeros(n)
for i in range(n):
#Si la frecuncia del punto no está dentro de la banda especificada, su valor es cero
    if(frecuencias[i]<frec_min or frecuencias[i]>frec_max):
        Y_filtrada[i]=0
    else:
        Y_filtrada[i]=Y[i]
#Transformada inversa de Fourier
y_filtrada=ifft(Y_filtrada)
```

```python
#Gráfica de la señal de entrada y la señal filtrada
plot(t,y,c='b')
plot(t,y_filtrada,c='k')
xlabel('Tiempo (s)')
ylabel('Amplitud normalizada')
legend(['Senial original','Senial filtrada'])
```

```python
#Se genera un archivo wav que guarda la señal filtrada
format = audiolab.Format('wav')
salida = audiolab.Sndfile('chimpanzee_filtrado.wav','w',format,1,tasa_muestreo)
salida.write_frames(y_filtrada.real)
salida.close()
```

```python
#Se genera una señal amplificada para poderla escuchar mejor
factor_amplificacion=10
y_filtrada_amplificada=y_filtrada*factor_amplificacion

plot(t,y,c='b')
plot(t,y_filtrada_amplificada)
xlabel('Tiempo (s)')
ylabel('Amplitud')
legend(['Senial original','Senial filtrada amplificada'])
axis([0,8,-2.5,2.5])

format = audiolab.Format('wav')
salida = audiolab.Sndfile('chimpanzee_filtrado_amplificada.wav','w',format,1,tasa_muestreo)
salida.write_frames(y_filtrada_amplificada.real)
salida.close() 
```
