### PERIODO CICLO SOLAR

```python
%pylab inline
from scipy.interpolate import interp1d
from scipy.fftpack import fft, fftfreq
```

```python
#Carga el archivo
data=loadtxt('sparse_sample_monthrg.dat')
#Obtiene la forma del archivo
shp=shape(data)
#Número de datos
N=shp[0]
```

```python
#Definición de una sola escala de tiempo en años decimales
tiempo=zeros(N)
for i in range (N):
    tiempo[i]=data[i,0]+(data[i,1]/12)
#Arreglo de datos iniciales
datos=zeros(N)
for i in range(N):
    if(data[i,3]!= -99.0):
        datos[i]=data[i,3]
#Gráfica de datos iniciales corregidos
scatter(tiempo,datos)
```

```python
#Interpolación de los datos por el método que arrojó más cercanía a los valores iniciales
datos_nuevos=interp1d(tiempo,datos,kind='nearest')
#Definición de variable de tiempo para graficar polinomio de interpolación
t=size(tiempo)
#La resolución de la variable de tiempo permite dar pasos que contemplen todos los puntos originales
tiempo_nuevo=linspace(tiempo[0],tiempo[t-1],2500)
#Gráfica de datos originales y polinomio de interpolación
plot(tiempo_nuevo,datos_nuevos(tiempo_nuevo),c='r')
scatter(tiempo,datos)
axis([1600,2000,-10,290])
xlabel('Fecha en anios decimales')
ylabel('Promedio de manchas solares')
legend(['Interpolacion','Datos originales'])
```

```python
#Transformada de Fourier de los datos
fourier=fft(datos_nuevos(tiempo_nuevo))
#Arreglo con la frecuencia de cada punto
frecuencias=fftfreq(len(fourier),1)
```

```python
#Posiciones de los puntos cuyas frecuencias tienen mayor amplitud, ordenadas de manera descendente
c = (argsort(-abs(fourier)))
print 'Frecuencias con mayores amplitudes:',abs(frecuencias[c[:10]])
```

```python
#Se observa que el valor en la posición 1 es el que tiene sentido físico puesto que el anterior es cero
freq=abs(frecuencias[c[1]])
periodo_solar=1/(freq*3600*24)*365
print 'El ciclo solar tiene un periodo de aproximadamente', round(periodo_solar),'anios'
```
