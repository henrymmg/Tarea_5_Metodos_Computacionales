# Diametro de Circulos

```python
%pylab inline
from scipy.fftpack import fft, fftfreq
```

```python
#importo los datos de los puntos y circulos

data=loadtxt(open('BAO.dat','r'))

x = data[:,0]
y = data[:,1]
```

```python
#genero un histograma para visualizar mis circulos por medio de un imshow

x1,y1,z= histogram2d(x,y, bins=(135,135))
imshow(x1)
```

```python
#numero de datos totales de en mi matriz

n_datos = shape(x1)[1]
print n_datos
```

```python
#hago fourier sobre los datos en mi matriz

dt= 1.0/n_datos
frecuencia = fftfreq(n_datos, dt)
fourier = (fft(x1))
```

```python
#organizo los datos de mis frecuencias de menor a mayor para coger mis 10 valores max

valores = 0
lista = multiply(-1, fourier)
argumento = argsort(lista)
maximos = empty(0)

for i in range(n_datos-1):
    maximos = append(maximos, fourier[argumento[i]])
maximos[:10]
```
